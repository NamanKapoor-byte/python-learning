```python
class Expense_tracker:
    def __init__(self,filename="expenses.txt"):
        self.filename=filename
    def add_expenses(self,items,amount):
        file=open(self.filename,"a")
        file.write(items+","+str(amount)+"\n")
        file.close()
        print("items :",items,"amount :",amount)
    def total_expenses(self):
        total=0
        file=open(self.filename,"r")
        for line in file:
            data=line.split(",")
            total=total + float(data[1])
        file.close()
        print("total money expend $ :",total)
    def clear_history(self):
        # Opening a file in "w" mode instantly wipes it completely clean
        file = open(self.filename, "w")
        file.close()
        print("History cleared successfully! Your tracker is reset to 0.")
tracker=Expense_tracker()
tracker.clear_history()
tracker.add_expenses("books",30)
tracker.add_expenses("Lunch",40)
tracker.total_expenses()

        
```
