```python
#oops concepts
class Phone:
    def set_color(self,color):
        self.color=color

    def set_cost(self,cost):
        self.cost=cost
    
    def show_color(self):
        return self.color
    
    def show_cost(self):
        return self.cost
    
    def make_call(self):
        print("lets make a phone call.")

    def play_games(self):
        print("lets play the games in phone.")
p1 = Phone()
p1.set_color("blue")
print(p1.show_color())
p1.set_cost(55000)
print(p1.show_cost())
p1.play_games()
p1.make_call()
print(" ")
# work with constructor

class Employee:
    def __init__(self,name,gender,salary,age):
        self.name=name
        self.gender=gender
        self.salary=salary
        self.age=age

    def show_details(self):
        print("name of the employee is :",self.name)
        print("gender of the employee is :",self.gender)
        print("salary of the employee is :",self.salary)
        print("age of the employee is :",self.age)
em = Employee("naman","male",55000,18)
em.show_details()
print(" ")
# inheritnace in python
class Vehicle:
    def __init__(self,mileage,cost):
        self.mileage=mileage
        self.cost=cost
    def show_details(self):
        print("I am vehicle.")
        print("mileage of vehicle :",self.mileage)
        print("cost of vehicle :",self.cost)
vh=Vehicle(300,10000)
vh.show_details()
print(" ")
class Car(Vehicle):                      # child class 
    def __init__(self,mileage,cost,tyres,hp):
        Vehicle.__init__(self,mileage,cost)              # overiding __init__function 
        self.tyres=tyres
        self.hp=hp
    
    def show_details(self):
        super().show_details()                  # can use Vehicle.show_details(self) also instead of super().
        print(" ")
        print("I am car.")
        print("tyres is :",self.tyres)
        print("hp is :",self.hp)
c1=Car(1200,300000,4,150)
c1.show_details()

# multiple inheritances (two parents)
print(" ")
class Vehicle:
    def __init__(self,mileage):
        self.mileage=mileage
class Electric_battery:
    def __init__(self,battery_capacity):
        self.battery_capacity=battery_capacity
class Electric_vehicle(Vehicle,Electric_battery):
    def __init__(self,mileage,battery_capacity,battery_period,brand):
        self.brand=brand
        self.battery_period=battery_period
        Vehicle.__init__(self,mileage)
        Electric_battery.__init__(self,battery_capacity)
    def show_details(self):
        print("Mileage of car :",self.mileage,"KM")
        print("Battery capacity of the car :",self.battery_capacity,"KMW")
        print("battery period of car :",self.battery_period,"YRS")
        print("brand of the car :",self.brand)

ev=Electric_vehicle(200,10000,7,"TESLA")
ev.show_details()

# second example of multiple inheritance (parent1,parent2.child)
class Parent1:
    def assign_str1(self,str1):
        self.str1=str1

    def show_str1(self):
        print("First :",self.str1)

class Parent2:
    def assign_str2(self,str2):
        self.str2=str2

    def show_str2(self):
        print("Second :",self.str2)

class Child(Parent1,Parent2):
    def assign_str3(self,str3):
        self.str3=str3

    def show_str3(self):
        print("Third :",self.str3)
c1=Child()
print(" ")
c1.assign_str1("hey!")
c1.assign_str2("hey!!")
c1.assign_str3("hey!!!")
c1.show_str1()
c1.show_str2()
c1.show_str3()

## third example of multiple-level inheritances (parent--child--grandchild)
print(" ")
class Parent:
    def assign_name(self,name):
        self.name=name
    def show_name(self):
        print("Name :",self.name)
class Child(Parent):
    def assign_age(self,age):
        self.age=age
    def show_age(self):
        print("Age :",self.age)
class Grandchild(Child):
    def assign_gender(self,gender):
        self.gender=gender
    def show_gender(self):
        print("Gender :",self.gender)
G1=Grandchild()
G1.assign_name("naman")
G1.assign_age("18")
G1.assign_gender("male")
G1.show_name()
G1.show_age()
G1.show_gender()


```
