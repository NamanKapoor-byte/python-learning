```python
class PCM_students:
    def __init__(self,name,physics,chemistry,math):
        self.name=name
        self.physics=physics
        self.chemistry=chemistry
        self.math=math
    def calculate_results(self):
        total=self.physics + self.chemistry + self.math
        percentage=(total/300)*100
        if percentage >=33:
            status="PASSED"
        else:
            status="FAILED"
        print("Student Name :",self.name)
        print("Percentage :",round(percentage,2))
        print("Status :",status)
student1=PCM_students("naman",85,88,92)
student1.calculate_results()


```
